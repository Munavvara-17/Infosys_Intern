```python
import matplotlib.pyplot as plt
```

```python
# simple line plot

x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.plot(x,y) # plotting line graph
plt.title("simple line grapgh") #title of graph
plt.xlabel("X-axis") #label of x-axis
plt.ylabel("Y-axis") #label of y-axis
plt.show() # show the plot
```

```python
#adding color, style and markers

x = [1,2,3,4,5]
y = [2,3,4,5,6]

plt.plot(x,y,color='red',linestyle='--',marker='o')
plt.title("simple line grapgh")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#bar chart

students = ["A","B","C","D","E"]
marks = [85,90,75,60,95]

plt.bar(students,marks,color ='brown')
plt.title("Students marks comparison")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
#horizontal bar chart

students = ["A","B","C","D","E"]
marks = [85,90,75,60,95]

plt.barh(students,marks,color ='green')
plt.title("Students marks comparison")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

```

```python
#scatter plot
import numpy as np
x = np.random.randint(1,10,10)
y = np.random.randint(1,10,10)

plt.scatter(x,y,color='blue',marker='*')
plt.title("Random scatter plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# TASK
import matplotlib.pyplot as plt
import numpy as np

students = ["A", "B", "C", "D", "E"]
marks = [85, 90, 75, 60, 95]
study_hours = [5, 6, 4, 3, 7]



# 1. Line Plot
plt.plot(students, marks, color='red', linestyle='--', marker='o')
plt.title("Marks Trend of Students (Line Plot)")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

# 2. Bar Chart
plt.bar(students, marks, color='brown')
plt.title("Students Marks Comparison (Bar Chart)")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

# 3. Horizontal Bar Chart
plt.barh(students, marks, color='green')
plt.title("Students Marks Comparison (Horizontal Bar)")
plt.xlabel("Marks")
plt.ylabel("Students")
plt.show()

# 4. Scatter Plot
plt.scatter(study_hours, marks, color='blue', marker='*')
plt.title("Study Hours vs Marks (Scatter Plot)")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
#pie chart
activities = ["study", "sleep", "Games", "exercise0"]
time = [8,7,4,5]

plt.pie(time, label = activities, autopct = "1.1f%%")
plt.title("Daily Activities Breakdown")
plt.show()
```

```python
#multiple plots in one figure

x = [1,2,3,4,5]
y1 = [1,4,6,7,8]
y2 = [25,20,15,10,5]

plt.subplot(1,2,1) # 1 row, 2 column, first plot
plt.plot(x, y1, color= 'blue')
plt.title("Increasing Trend")

plt.subplot(1,2,2) # 1 row, 2 column, second plot
plt.plot(x, y2, color= 'red')
plt.title("Decreasing Trend")

plt.show()
```

```python
#adding legends

x = [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [3,6,9,12,15]

plt.plot(x,y1,label = "Even Series", color = "blue")
plt.plot(x,y2,label = "Odd Series", color="red" )
plt.title("line grapgh with legend")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()   #show the legend
plt.show()
```

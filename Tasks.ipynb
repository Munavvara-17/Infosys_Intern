```python
# user input and check number is +ve, -ve or zero
num = float(input("Enter a number: "))
if num > 0:
    print("The number is positive.")    
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
```

```python
# function that return square of a number

def square(num):
    return num ** 2

# take input from user
number = int(input("Enter a number: "))

# call function
result = square(number)

print("Square is:", result)
```

```python
# count how many times each character appears in a string
def count_characters(string):
    char_count = {}
    for char in string:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    return char_count

# take input from user
text = input("Enter a string: ")

# call function
result = count_characters(text)

print("Character count:", result)
```

```python
# create dataframe and find avg salary
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],    
    'Salary': [50000, 60000, 70000, 80000, 90000]
}
df = pd.DataFrame(data)
avg_salary = df['Salary'].mean()
print(f"Average Salary: {avg_salary}")
```

```python
# find missing number in a sequence 
nums = [1, 2, 3, 5]

n = len(nums) + 1
expected_sum = n * (n + 1) // 2
actual_sum = sum(nums)

if expected_sum != actual_sum:
    print("Missing number:", expected_sum - actual_sum)
else:
    print("No number missing")
```

```python
# reverse word in a string without split
s = input("Enter a string: ")
word = ""
result = ""
for char in s:
    if char != " ":
        word = char + word
    else:
        result += word + " "
        word = ""   
print(result + word)
```

```python
# without loop a num is power of 2

n = int(input("Enter a number: "))
if n > 0 and (n & (n - 1)) == 0:
    print("is a power of 2.")
else:
    print(" is not a power of 2.")
```

```python
# flatten a list without using loop

lst = [1, [2, 3], [4, 5]]
flat = []

for item in lst:
    if type(item) == list:
        for i in item:
            flat.append(i)
    else:
        flat.append(item)

print(flat)
```

```python
# find first non repeating character without loop

s = "aabbcdd"

for ch in s:
    if s.count(ch) == 1:
        print("Non-repeating character:", ch)
        break
```

```python
# Check if Two Strings Are Anagrams

str1 = input("Enter the first string: ")
str2 = input("Enter the second string: ")

if sorted(str1) == sorted(str2):
    print("anagrams")
else:
    print("not anagrams")
```

```python
# ,Count Frequency of Each Word

word = input("enter a word:")
freq = {}

for i in word:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1

for char, count in freq.items():
    print(char, count)
```

```python
w = input("Enter a sentence: ")

words = w.split()          # split sentence into words
reversed_words = words[::-1]   # reverse word list

result = " ".join(reversed_words)  # join back into sentence

print("Reversed sentence:", result)
```
